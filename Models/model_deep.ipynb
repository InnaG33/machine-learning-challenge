```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
len(df)
```

```python
df['koi_disposition'].value_counts()
```

```python
columns = df.keys()
columns
```

# Select your features (columns)

```python
list_features = columns[1:]
list_features
```

```python
selected_features = df[list_features]
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
y = df["koi_disposition"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

```python
len(X_train)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_scaler = MinMaxScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

# Train the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=10, activation='relu', input_dim=40))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

# Evaluate and Predict

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
predictions = model.predict_classes(X_test_scaled)
```

```python
encoded_results = label_encoder.inverse_transform(predictions)
```

```python
deepmodel_df = pd.DataFrame({"Prediction": encoded_results, "Actual": y_test})
deepmodel_df.reset_index(inplace=True, drop=True)
deepmodel_df
```

# Save the Model

```python
# dm_train_score=model.score(X_train_scaled, y_train_categorical)
# dm_test_score=model.score(X_test_scaled, y_test_categorical)
deepmodel_acc = pd.DataFrame({"model_loss": [model_loss], "model_accuracy": [model_accuracy]})
deepmodel_acc
```
